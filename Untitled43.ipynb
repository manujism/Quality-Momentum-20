```python
import pandas as pd
import yfinance as yf
from collections import defaultdict
import warnings
from IPython.display import display

warnings.simplefilter(action='ignore', category=FutureWarning)

# Load databank.csv
df = pd.read_csv("databank.csv")

df.rename(columns=lambda x: x.strip(), inplace=True)

# Apply filters based on screenshot data
df = df[df['Debt to equity'] <= 1.5]
df = df[df['Sales growth 3Years'] > 0]
df = df[df['Average return on equity 3Years'] >= 10]
df = df[df['Piotroski score'] >= 5]
df = df[~df['Ticker'].str.upper().str.contains('ADANI')]

# Prepare ticker list and sector map
tickers = df['Ticker'].tolist()
sector_map = df.set_index('Ticker')['Sector'].to_dict()

# Download price data
start_date = '2022-01-01'
end_date = '2025-06-01'
raw_data = yf.download(tickers, start=start_date, end=end_date, interval="1mo")


if 'Close' in raw_data.columns:
    data = raw_data['Close']
else:
    data = raw_data.xs('Close', level=1, axis=1)

data.index = pd.to_datetime(data.index)
data.index.name = 'Date'


momentum_3m = (data / data.shift(3)) - 1
momentum_6m = (data / data.shift(6)) - 1
momentum_12m = (data / data.shift(12)) - 1

z_3m = (momentum_3m - momentum_3m.mean()) / momentum_3m.std()
z_6m = (momentum_6m - momentum_6m.mean()) / momentum_6m.std()
z_12m = (momentum_12m - momentum_12m.mean()) / momentum_12m.std()


combined_score = 0.4 * z_6m + 0.4 * z_12m + 0.2 * z_3m
z_final = (combined_score - combined_score.mean()) / combined_score.std()
z_final_winsorized = z_final.clip(lower=-3, upper=3)

momentum_score = z_final_winsorized.applymap(lambda x: 1 + x if x > 0 else (1 - x)**-1)
latest_scores = momentum_score.iloc[-1]
sorted_scores = latest_scores.sort_values(ascending=False)


sector_count = defaultdict(int)
max_per_sector = 4
top_stocks = []

for ticker in sorted_scores.index:
    sector = sector_map.get(ticker, 'Unknown')
    if sector_count[sector] < max_per_sector:
        top_stocks.append(ticker)
        sector_count[sector] += 1
    if len(top_stocks) == 20:
        break


final_df = df.set_index('Ticker').loc[top_stocks].reset_index()
final_df['Score'] = latest_scores[top_stocks].values
display(final_df[['Ticker', 'Sector', 'Debt to equity', 'Sales growth 3Years',
                  'Average return on equity 3Years', 'Piotroski score', 'Score']])
```
